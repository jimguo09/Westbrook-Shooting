### Setting Up the Investigation

This investigation therefore looks at two inquiries:


<li>  1. Is there convincing evidence to show Westbrook's 21-22 season 3pts percent is statistically significant relative to his 3pts percent entering the season?
<ul>
<li> <small> In other words, is his season 3pts percent is uncharacteristically lower than his 3pts percent entering the season? </small></li>
</ul>
<li> 2. Is there convincing evidence that Westbrook's 21-22 season 3pts percent is worse than the league average 3pts percent?

\

Since we will be investigating claims about the true percentages, which qualifies as proportions, of a population, that is Westbrook's total 3pts attempts, we will be electing to use a z-test, a form of hypothesis testing. Furthermore, given our two inquiries, in which both only uses one sample that his is 21-22 total 3ptss attempted, we will be using two one-sample z-test for the total of this investigation.

\

For both inquiries that necessitates hypothesis testing, we wish to test Ho: p = Westbrook's 3pts percentage entering the 21-22 season or league average 3pts percentage vs Ha: p < Westbrook's 3pts percentage entering the 21-22 season or league average 3pts percentage, given that p = Westbrook's true percentage of 3pts makes — the parameter of interest — using a significance level of 0.05. This will be stated specifically under the inference section.

### Key Assumptions Made to Reduce Confounding Bias

This investigation made two assumptions to reduce confounding biases:

1. In this study, we will limit Westbrok's career from his 2010-11 season to his 2021-22 season. This is because his emergence as a star player began in the 10-11 season, where he captured his first All-Star team nomination from both fans, league executives, media members, and analysts. Effectively, this study studies Westbrook's 3pts shooting since he first broke-out as a top player. Thus, if including the prior two seasons in which Westbrook was still developing as a raw prospect, it could result in confounding biases that may lower Westbrook's 3pts percentage entering the 21-22 season, which is inappropriate within the context of this study. Hence, we will now effectively refer Westbrook's *career* in this study from the 2010-11 season to the 2021-2022 season.
2. This is mentioned later under the specific inferences section, but we will assume that the result of making one 3pts shot does not change the probability to making the next. In other words, we assume that each 3pts attempt is independent of each other. In the real world, though, this is often not the case. Players often get "hot" and can get "cold;" it is a known trend that players can go on "streaks" of making 3ptss in a row or missing all their attempts in a game. Though, for this study, we will ignore this phenomenon as it is a confounding random factor: it is often unknown when players would go on these "streaks" and would force players to take much more or much less of shots they are suppose to take. Having more of them in a particular season could confound a player's 3pts percentage, and thus we will assume that each 3pts attempt is independent.

### Deriving Needed Statistics for Hypothesis Tests from Westbrook's 3pts Shooting Data:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as py

# Restate the csv-imported data:

rawData = {"2010-11":[82,0.4,1.3],
           "2011-12" :[66,0.9,3.0],
           "2012-13":[82,1.2,3.7],
           "2013-14":[46,1.5,4.7],
           "2014-15":[67,1.3,4.3],
           "2015-16" :[80,1.3,4.3],
           "2016-17": [81,2.5,7.2],
           "2017-18":[80,1.2,4.1],
           "2018-19" :[73,1.6,5.6],
           "2019-20" : [57,1.0,3.7],
           "2020-21" :[65,1.3,4.2],
           "2021-22" : [78,1.0,3.4]}

## rawData is a dictionary representation of the csv raw data of Westbrook's
# 3 point shooting from the 2010-11 season to 2021-22 season.

## The dictionary is in the form of key : value, where key is a string representation
# of the NBA regular season and the value is an array of Westbrook's shooting data
# in the form of [games played, 3pts made/game, 3pts attempted/game]

## Calculate Westbrook's total 3pts made and 3pts attempted from 2010-11 to 2021-22:
tot3Made = 0
tot3Att = 0
for season in rawData.keys():
  seasonStats = rawData[season]
  seasonTot3Made = seasonStats[0] * seasonStats[1]
  seasonTot3Att = seasonStats[0] * seasonStats[2]
  tot3Made += seasonTot3Made
  tot3Att += seasonTot3Att

tot3Made = round(tot3Made, 0)
tot3Att = round(tot3Att, 0)
print("Westbrook's career 3pts made: " + str(tot3Made))
print("Westbrook's career 3pts attempt: " + str(tot3Att))

## Calculate Westbrook's total 3pts made and 3pts attempted in the 2021-22 season:
thisSeason = rawData["2021-22"]
season3Made = round(thisSeason[0] * thisSeason[1], 0)
season3Att = round(thisSeason[0] * thisSeason[2], 0)

print("Westbrook's total 3pts made in the 2021-2022 season: " + str(season3Made))
print("Westbrook's total 3pts attempted in the 2021-2022 season: " + str(season3Att))

## Calculating Westbrook's total 3pts made and 3pts attempted entering 2021-22 season
# or from 2010-11 season to 2020-21 season:

beforeThisSeason3Made = round(tot3Made - season3Made, 0)
beforeThisSeason3Att = round(tot3Att - season3Att, 0)

print("Westbrook's total 3pts made entering the 2021-2022 season: " + str(beforeThisSeason3Made))
print("Westbrook's total 3pts attempted entering the 2021-2022 season: " + str(beforeThisSeason3Att))

## Caculating 3pts percentages
tot3Percent = round(tot3Made / tot3Att, 4)
season3Percent = round(season3Made / season3Att, 4)
beforeThisSeason3Percent = round(beforeThisSeason3Made / beforeThisSeason3Att, 4)
print("Westbrook's career 3pts percentage: " + str(tot3Percent))
print("Westbrook's 2021-2022 season 3pts percentage: " + str(season3Percent))
print("Westbrook's career 3pts percentage entering the 2021-2022 season: " + str(beforeThisSeason3Percent))


```

Note that for the sake of the study, we will consider the number of 3pts attempted and made in 21-22 season by Westbrook, or 265 attempts, as a random sample of 3ptss attempted and made. Subsequently, Westbrook's 21-22 season 3pts percentage will be the considered the "sample proportion."

\
The justification is that if a random sample from a simulation, say using a random number generator, was drawn using the 21-22 data, that proportion of 3ptss made will be inappropriate since we already know Westbrook's 21-22 3pts percentage. Thus, we consider the 21-22 sample of 3pts shots attempted and made by Westbrook as a random sample of his career 3pts made and attempted.

## Inferences

We will conduct two one sample z-test for our inference procedures:

### Is There Convincing Evidence to Show Westbrook's 21-22 3pts Percent is Statistically Significant Relative to his 3pts Percent Entering the Season?

\

For context, Westbrook's 3pts percentage entering 21-22 season is 0.3092; the 3pts percentage that Westbrook had in the 21-22 season is 0.2943; the number of shots he attempted in the 21-22 season is 265; and the total shots he attempted throughout his career entering the 21-22 season is 3260.

\

Thus, we wish to test Ho: p = 0.3092 vs Ha: p < 0.3092, where p is the true proportion of 3 point shots that Westbrook makes using α = 0.05.

\

We will conduct a one sample z-test for p. First, check the conditions to satisfy a z-test:

*   Randomization: Westbrook attempted 265 3pts shots in the 21-22 season. The sample is independent assuming that the result of Westbrook making one three pointer will not affect the probability of him making the next. The sample is also random, given that the 265 shots attempted in this season can be thought of a random sample of the total number of shots Westbrook took throughout his career.
*   10% Condition: 265 3pts shots attempted is 10% of Westbrook's total 3 point shots attempted, which is 3525.
*   Large Counts Condition: 100(0.3092) = 30.92 >= 10, and 100(1-0.3092) = 69.08 >= 10.

\

All the conditions for the test have been satisfied; Compute the z-statistic:





```python
import math
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import scipy
from matplotlib.ticker import FormatStrFormatter

#Compute the z-statistic
z = (0.2943 - 0.3092)/ math.sqrt(0.3092 * (1-0.3092) / 265)
print("This is the z-statistic: " + str(z))

print()
print()
print("Let the z-statistic be represented on a standard normal distribution curve: ")
# Graphically show the z-statistic on a default normal curve:
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = plt.axvline(x=z, ymin= 0, ymax = 1)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
ax1 = plt.gca()
ax1.get_yaxis().set_visible(False)
plt.xlabel('z-statistic')
plt.show()

#Use the scipy.sf function to technilogically produce a p-value given the z statistic.
pVal = scipy.stats.norm.sf(abs(z))
print()
print("The p value is essentially the area enclosed by the standard normal distribution curve \n with a mean of 0 and a standard deviation of 1 and the said z-statistic represented as a vertical line. \n The p value is calculated as: " + str(pVal))
```
